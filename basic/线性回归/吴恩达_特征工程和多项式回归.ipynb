使用线性回顾拟合复杂的情况，甚至是非线性的函数

```python
import matplotlib.pyplot as plt
import numpy as np
from lab_utils_multi import zscore_normalize_features, run_gradient_descent_feng

np.set_printoptions(precision=2)
```

```python
x = np.arange(0, 20, 1)
y = 1 + x ** 2
X = x.reshape(-1, 1)
model_w, model_b = run_gradient_descent_feng(X, y, 1000, 1e-2)

plt.scatter(x=x, y=y, marker='x', c='r', label='actual value')
plt.title("no feature engineering")
plt.plot(x, X @ model_w + model_b, label="Predicted Value")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
x = np.arange(0., 20, step=1)
y = 1 + x ** 2
X = x ** 2
X = X.reshape(-1, 1)
#  把所有的 x 转换为 x**2
model_w, model_b = run_gradient_descent_feng(X, y, 10000, 1e-6)

plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.title("Added x**2 feature")
plt.plot(x, np.dot(X, model_w) + model_b, label="Predicted Value")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
x = np.arange(0, 20, 1)
y = x ** 2
X = np.c_[x, x ** 2, x ** 3]
```

```python
model_w, model_b = run_gradient_descent_feng(X, y, 10000, 1e-7)
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.title("x, x**2, x**3 features")
plt.plot(x, X @ model_w + model_b, label="Predicted Value")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
